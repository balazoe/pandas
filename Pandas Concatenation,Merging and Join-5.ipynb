### import Necessary modules 

```python
import pandas as pd
```

```python
north_america = pd.read_csv('north_america_2000_2010.csv',index_col=0)
north_america
```

```python
south_america = pd.read_csv('south_america_2000_2010.csv',index_col=0)
south_america
```

### create line grap for our yearly labor trends in north_america

```python
import matplotlib.pyplot as plt
```

```python
north_america.plot()
```

### Plot transposed line grap of north_america dataframe , with title "Average Labor Hours per Year"

```python
north_america.transpose().plot(title='Average Labor Hours Per Year')
```

### Plot transposed line grap of south_america dataframe , with title "Average Labor Hours per Year"

```python
south_america.transpose().plot(title='Average Labor Hours Per Year')
```

### Concatate north_america and south_america dataframes and store results in a dataframe

```python
americas = pd.concat([north_america,south_america])
americas
```

### Load additional files

```python
americas_dfs = [americas]
for year in range(2011,2016):
    filename = "./americas_{}.csv".format(year)
    df = pd.read_csv(filename, index_col=0)
americas_dfs.append(df)
```

```python
americas_dfs[1]
```

### Concatenate americas and americas_dfs dataframes and store results in americas

```python
americas = pd.concat(americas_dfs,axis=1)
```

```python
americas.index.names = ['Country']
```

```python
americas
```

### Plot transposed americas dataframe

```python
americas.transpose().plot(title='Average Labor Hours Per Year')
```

### Appending data from other Continents

```python
asia = pd.read_csv('asia_2000_2015.csv',index_col=0)
asia
```

```python
europe = pd.read_csv('europe_2000_2015.csv',index_col=0)
europe.head()
```

```python
south_pacific = pd.read_csv('south_pacific_2000_2015.csv',index_col=0)
south_pacific
```

### Append asia europe and south_pacific to americas dataframe and assign to new dataframe world

```python
world = americas.append([asia,europe,south_pacific])
```

```python
world.index
```

### Append transposed world dataframe 

```python
world.transpose().plot(title='Average Labor Hours Per Year')
```

### Customize this plot 

```python
world.transpose().plot(figsize = (10,10),colormap = 'rainbow',linewidth=2,title='Average Labor Hours Per Year')
plt.legend(loc='right')
plt.show()
```

### Merging Historical Labor data

```python
historical = pd.read_csv('historical.csv',index_col=0)
historical.head()
```

```python
print("world rows & columns:",world.shape)
print("historical rows & columns:",historical.shape)
```

### Merge historical dataframe with world dataframe as world_historical

```python
world_historical =  pd.merge(historical,world,left_index=True,right_index=True,how='right')
```

```python
world_historical.shape
```

```python
world_historical.head()
```

### Joining Historical Data

```python
world_historical = historical.join(world,how='right')
```

```python
world_historical.head()
```

### Plot the world labor data

```python
world_historical.sort_index(inplace=True)
```

```python
world_historical.transpose().plot(figsize = (10,10),colormap = 'rainbow',linewidth=2,title='Average Labor Hours Per Year')
plt.legend(loc='right',bbox_to_anchor=(1.5,0.5))
plt.show()
```

### which country worked larger hours per year? 

```python
world_historical.index.name='country'
```

```python
world_historical.groupby('country').max()
```

### which country worked Shorter hours per year? 

```python
world_historical.groupby('country').min()
```
